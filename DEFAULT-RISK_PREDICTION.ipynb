```python
import pandas as pd
import numpy as np
```

```python
dt = pd.read_csv('vedika_final.csv')
dt.head()
```

```python
dt = dt.drop_duplicates(keep='last')
```

```python
dt.info()
```

```python
dt.head()
```

```python
dt.drop(['SI No', 'Member No'], axis = 1, inplace = True)
```

```python
dt.drop(['Spouse Employment  status - Employed/unemployed','Duration of credit'], axis = 1, inplace = True)
dt.head()
```

```python
dt.drop(['Gender', 'Marital status'], axis = 1, inplace = True)
```

```python
dt.head()
```

```python
dt.drop(['No of dependent family members'], axis = 1, inplace = True)
```

```python
dt.drop(['Employment status - Temporary/Permanent or Employed/Unemployed '], axis = 1, inplace = True)
```

```python
dt.head()
```

```python
dt.drop(['Unnamed: 0'], axis = 1, inplace = True)
```

```python
dt.head()
```

```python
x = dt.drop(['Paid /Non Paid (Default)'], axis = 1)
```

```python
y = dt['Paid /Non Paid (Default)']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x['Educational level - literate/Illiterate'] = le.fit_transform(x['Educational level - literate/Illiterate'])
```

```python
le2 = LabelEncoder()
x['Nature of livelihood - Farm/Non Farm'] = le2.fit_transform(x['Nature of livelihood - Farm/Non Farm'])
```

```python
x.head()
```

```python
le3 = LabelEncoder()
y = le3.fit_transform(y)
```

```python
y
```

```python
x['Monthly Income per family member'] = x['Monthly Income']/x['No of Family Members']
```

```python
x.drop(['Monthly Income','No of Family Members'], axis = 1, inplace = True)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, test_size=0.2, random_state=42)
```

```python
x['Sector of livelihood'].isna().sum()
```

```python
x_train['Sector of livelihood'].isna().sum()
```

```python
x_train['Sector of livelihood'].value_counts
```

```python
x_train['Sector of livelihood'].fillna(x_train['Sector of livelihood'].mode()[0], inplace = True)
```

```python
x_train['Sector of livelihood'].isna().sum()
```

```python
x_test['Sector of livelihood'].isna().sum()
```

```python
x_test['Sector of livelihood'].fillna(x_test['Sector of livelihood'].mode()[0], inplace = True)
x_test['Sector of livelihood'].isna().sum()
```

```python
x_train
```

```python
x_test
```

```python
ordinal_label = {k : i for i, k in enumerate(x_train['Sector of livelihood'].unique(),0)}
x_train['Sector of livelihood'] = x_train['Sector of livelihood'].map(ordinal_label)
```

```python
ordinal_label2 = {l : j for j, l in enumerate(x_train['Purpose of credit'].unique(),0)}
x_train['Purpose of credit'] = x_train['Purpose of credit'].map(ordinal_label2)
```

```python
ordinal_label = {k : i for i, k in enumerate(x_test['Sector of livelihood'].unique(),0)}
x_test['Sector of livelihood'] = x_test['Sector of livelihood'].map(ordinal_label)
```

```python
ordinal_label2 = {l : j for j, l in enumerate(x_test['Purpose of credit'].unique(),0)}
x_test['Purpose of credit'] = x_test['Purpose of credit'].map(ordinal_label2)
```

```python
from sklearn.linear_model import LogisticRegression
c1 = LogisticRegression(random_state=0)
c1.fit(x_train, y_train)
```

```python
y_pred = c1.predict(x_test)
y_pred_train = c1.predict(x_train)
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
cm2 = confusion_matrix(y_train, y_pred_train)
acc1 = accuracy_score(y_test, y_pred)
acc2 = accuracy_score(y_train, y_pred_train)
print(acc1, acc2)
```

```python
from sklearn.metrics import plot_confusion_matrix
disp = plot_confusion_matrix(c1, x_test, y_test, cmap='Blues', values_format='.3g')
```

```python
from sklearn.ensemble import RandomForestClassifier
c6 = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
c6.fit(x_train, y_train)
```

```python
y_pred = c6.predict(x_test)
y_pred_train = c6.predict(x_train)
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
cm2 = confusion_matrix(y_train, y_pred_train)
acc1 = accuracy_score(y_test, y_pred)
acc2 = accuracy_score(y_train, y_pred_train)
print(acc1, acc2)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=c6, X=x_train, y=y_train, cv=10)
```
